```python
x = np.linspace(1, 10, 10)
x
```

```python
x = np.linspace(-np.pi, np.pi, 50)
```

```python
y = x
```

```python
X, Y = np.meshgrid(x, y)
```

```python
func = lambda x, y: np.cos(y) / (1 + x**2)
```

```python
Z = func(X, Y)
```

```python
plt.contour(X, Y, Z)
```

```python
plt.contour(X, Y, Z, levels=45, corner_mask=True)
```

```python
fa = (Z - Z.T) / 2
```

```python
plt.contour(X, Y, fa, levels=15)
```

```python
plt.contourf(X, Y, Z)
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z)
ax.view_init(30, 20)
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z)
ax.view_init(30, 70)
```
